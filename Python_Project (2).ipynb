```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Initial Public Offering.xlsx")
df.head()
```

# #check duplicate rows (entire row)

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

# #Drop specific unnamed columns

```python
# Drop specific unnamed columns
df = df.drop(columns=[
    'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
    'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'
], errors='ignore')
```

```python
df.head()
```

```python
df.info()
```

## # Replace missing numeric values with mean.

```python
num_cols = [
    'Issue_Size(crores)', 
    'Listing Gain', 
    'QIB', 
    'HNI', 
    'RII', 
    'Total'
]

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df
```

## # Calculate and display the number of IPOs for each month.

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month_name()

monthly_ipo = df.groupby('Month').size().sort_values(ascending=False)

print(monthly_ipo)
```

```python
top_10_ipos = df.sort_values(
    by='Issue_Size(crores)', 
    ascending=False
).head(10)

top_10_ipos[['IPO_Name', 'Issue_Size(crores)']]
```

## # Display the top 3 months with the highest number of IPOs.

```python
# Top 3 months
top_3_months = monthly_ipo.head(3)
print("Top 3 IPO months:\n", top_3_months)
```

## # Identify top 10 IPOs by issue size and calculate average listing gain.

```python
top10_issue = df.sort_values(
    'Issue_Size(crores)', ascending=False
).head(10)

avg_listing_gain = top10_issue['Listing Gain'].mean()

top10_issue['Listing Gain']
print("Overall Avg Listing Gain:", avg_listing_gain)
```

## # Analyze average subscription levels across IPO issue size categories.

```python
bins = [0, 500, 1500, 3000, df['Issue_Size(crores)'].max()]
labels = ['≤500', '501–1500', '1501–3000', '>3000']

df['Size Bucket'] = pd.cut(df['Issue_Size(crores)'], bins=bins, labels=labels)

bucket_analysis = df.groupby('Size Bucket')[[
    'QIB',
    'HNI',
    'RII',
    'Total'
]].mean()

print(bucket_analysis)
```

## # Analyze average listing gain across different subscription levels.

```python
subscription_listing_analysis = (
    df.groupby('Size Bucket')['Listing Gain']
      .mean()
)

print(subscription_listing_analysis)
```

## # Identify IPOs with negative listing gains.

```python
average_listing_gain = df['Listing Gain'].mean()
print("Average Listing Gain:", average_listing_gain)

negative_listing_ipos = df[df['Listing Gain'] < 0]

print("\nIPOs with Negative Listing Gain:")
print(negative_listing_ipos[['IPO_Name','Listing Gain']])
```

## # Find IPOs common to top subscription and top listing gain categories.

```python
top10_subscription = df.sort_values(
    'Total', ascending=False
).head(10)

top10_listing = df.sort_values(
    'Listing Gain', ascending=False
).head(10)

common_ipos = set(top10_subscription['IPO_Name']).intersection(
    set(top10_listing['IPO_Name'])
)

print("Common IPOs:", common_ipos)
print("Count:", len(common_ipos))
```

## # Calculate and display key IPO summary statistics.

```python
total_ipos = df['IPO_Name'].nunique()
print("Total IPOs:", total_ipos)
print()
avg_listing_gain = df['Listing Gain'].mean()
print("Average Listing Gain (%):", avg_listing_gain)
print()
avg_total_subscription = df['Total'].mean()
print("Average Total Subscription:", avg_total_subscription)
print()
positive_ipos = df[df['Listing Gain'] > 0].shape[0]
total_ipos_count = df.shape[0]

positive_ipo_percentage = (positive_ipos / total_ipos_count) * 100
print("Positive IPO Percentage:", positive_ipo_percentage)
```

## # Compare BSE and NSE closing market prices using a scatter plot.

```python
plt.scatter(df['CMP(BSE)'], df['CMP(NSE)'])
plt.xlabel("CMP (BSE)")
plt.ylabel("CMP (NSE)")
plt.title("CMP BSE vs CMP NSE Comparison")
plt.show()
```

```python
df['Price_Difference'] = df['CMP(NSE)'] - df['CMP(BSE)']
df[['IPO_Name', 'CMP(BSE)', 'CMP(NSE)', 'Price_Difference']].head()
```

```python
def better_exchange(diff):
    if diff > 0:
        return 'NSE'
    elif diff < 0:
        return 'BSE'
    else:
        return 'Same'

df['Better_Exchange'] = df['Price_Difference'].apply(better_exchange)

df[['IPO_Name', 'Better_Exchange']].head()
```

```python
exchange_count = df['Better_Exchange'].value_counts()

exchange_count.plot(kind='bar')
plt.title("Which Exchange Offers Better CMP?")
plt.xlabel("Exchange")
plt.ylabel("Number of IPOs")
plt.show()
```

# #WHICH INVESTOR CATEGORY DRIVES TOTAL SUBSCRIPTION MOST

```python
df['QIB'] = pd.to_numeric(df['QIB'], errors='coerce')
df['HNI'] = pd.to_numeric(df['HNI'], errors='coerce')
df['RII'] = pd.to_numeric(df['RII'], errors='coerce')
avg_sub = df[['QIB', 'HNI', 'RII']].mean()
avg_sub
plt.figure(figsize=(6,6))
plt.pie(
    avg_sub,
    labels=avg_sub.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Average IPO Subscription by Investor Type')
plt.show()
```

# #Listing Gain by Issue Size Bucket

# #WHICH INVESTOR TYPE CORRELATES MOST WITH LISTING GAIN

```python
cols = ['QIB', 'HNI', 'RII', 'Listing Gain']
corr = df[cols].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation: Investor Type vs Listing Gain')
plt.show()
```

# #NEGATIVE LISTING GAIN FREQUENCY

```python
neg_count = (df['Listing Gain'] < 0).sum()
pos_count = (df['Listing Gain'] >= 0).sum()

plt.figure(figsize=(6,4))
plt.bar(['Positive / Zero Gain', 'Negative Gain'], [pos_count, neg_count])
plt.title('IPO Listing Outcome Distribution')
plt.ylabel('Number of IPOs')
plt.show()
```
